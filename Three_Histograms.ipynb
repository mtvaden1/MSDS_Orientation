```python
import numpy as np
import matplotlib.pyplot as plt
```

### Binomial Distribution

```python
b = np.random.binomial(100, .5, 100)

plt.hist(b)
plt.show()
```

### Normal Distribution

```python
n = np.random.normal(100, 10, 100)

plt.hist(n)
plt.show()
```

### Chi-Squared Distribution

```python
c = np.random.chisquare(100, 100)

plt.hist(c)
plt.show()
```
