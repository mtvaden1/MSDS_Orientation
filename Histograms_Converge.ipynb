## Distributions converge as N increases:

### Binomial Distributions:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
r = [1, 5, 10, 50, 100, 500, 1000, 10000]

for n in r:
    binomial = np.random.binomial(n, .5, 1000)
    mu, std = norm.fit(binomial) 
  
    plt.hist(binomial, bins=15, density=True)
    
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 1000)
    p = norm.pdf(x, mu, std)
    
    plt.plot(x, p, 'k', linewidth=1)
    plt.title("n: {:.2f}".format(n))
    plt.show()
```

### Normal Distribution:

```python
print(mu, std)
# normal distribution with same mean and standard deviation as sufficiently large binomial distribution
normal = np.random.normal(mu, std, 1000)

plt.hist(normal, bins=15, density=True)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
p = norm.pdf(x, mu, std)
  
plt.plot(x, p, 'k', linewidth=1) 
plt.title("n: 10000")
plt.show()
```
